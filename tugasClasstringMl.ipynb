# Pengertian Clustring
clustring adalah setrategi pembelajaran tanpa pengawasan dari kumpulan data, lalu akan dimasukan kebebrapa kelompok/cluster.

# tipe clustring


1.  k-means

Algoritma k-mean merupakan algoritma unsupervised learning digunakan untuk pengelompokan data dan pengenalan pola. dengan memilih k = titik data awal secara acak, untuk menentukan pengelompokan secara ideal.

2. Hirarchical clustring

Hirarchical clustring adalah Pengelompokan berbasis konektivitas, di dasarkan prinsip bahwa setiap objek akan terhubung dengan tetangnya berdasarkan jarak terdekat(drajat hubungan). perlu di ingat bahwa clustring di represntasikan dengan Dendrogram, Sumbu X mewakili objek yg tidak tergabung dan sumbu Y jarak penggabung cluster.

3. Centroid-based or Partition Clustering(Pengelompokan Berbasis Centroid atau Partisi).

Pengelompokkan berbasis Centroid adalah pengelompokan bekerja berdasarkan kedekatan titik-titik dengan nilai pusat yang dipilih.

4. Density-based Clustring(Model-based Methods)

kalo diterjemahkan dalam bahasa indonesia "Pengelompokan berbasis Kepadatan(Metode berbasis model)" adalah metode yg mempertimbangkan densitas daripada jarak. maksudnya data di kelompokan berdasarkan wilayah dengan konsentrasi objek data yang tinggi dengan dibatasi oleh wilayah dengan konsetrasu objek data yg rendah.


# Urutan Algoritma K-means


1. Pemilihan Jumlah Klaster (K) dan Sentroid:

  Pilih jumlah klaster (K) yang diinginkan.

  Pilih sentroid awal untuk setiap klaster. Sentroid awal ini dapat dipilih secara acak dari titik-titik data dalam dataset.

2. Acak Data dan Inisialisasi Sentroid:

  Acak urutan titik-titik data dalam dataset.
  Inisialisasi sentroid yang telah dipilih pada langkah pertama.

3. Pemberian Klaster ke Titik Data:

  Hitung jarak antara setiap titik data dan setiap sentroid.
  Tentukan klaster untuk setiap titik data dengan mengaitkannya dengan sentroid terdekat.
  Setiap titik data akan diatributkan ke klaster yang memiliki sentroid terdekat dengan menggunakan aturan terdekat.

3. Pembuatan Sentroid Baru:

  Hitung rata-rata (mean) dari semua titik data yang termasuk dalam setiap klaster.

  Posisi baru ini akan menjadi sentroid yang diperbarui untuk setiap klaster.

4. Pengulangan Hingga Konvergensi:

  Ulangi langkah-langkah 3 dan 4 di atas sampai tidak ada perubahan signifikan dalam klaster atau sampai mencapai kriteria berhenti yang ditentukan, misalnya, jumlah iterasi maksimum atau ketika tidak ada perubahan dalam posisi sentroid.

Jadi pengimplematasian clustering pada mengidentifikasi basis pelanggan potensial untuk menjual produk dijadikan untuk pengambilan keputusan dalam menentuka strategi bisnis yg lebih efektif untuk meningkatkan penjualan peroduk kepada pelanggan yg tepat.

```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Masukkan data ke dalam DataFrame
data = {
    "Kabupaten" : ['kab bandung', 'kab bogor', 'kab cianjur', 'kab ciamis', 'kab sukabumi', 'kab majalengka', 'kab tasikmalaya', 'kab garut'],
    "Luas Lahan" : [65.500, 30.235, 55.65, 60.80, 75.89, 172.532, 78.90, 88.93],
    "Produksi Padi" : [405.056, 175.543, 259.78, 350.86, 568.92, 956.564, 653.321, 723.87]
}

df = pd.DataFrame(data)

# Lakukan one-hot encoding terhadap kolom "Kabupaten"
df_encoded = pd.get_dummies(df['Kabupaten'])

# Gabungkan DataFrame yang telah di-encoded dengan kolom "Luas Lahan" dan "Produksi Padi"
df_encoded[['Luas Lahan', 'Produksi Padi']] = df[['Luas Lahan', 'Produksi Padi']]

# Pilih jumlah klaster
k = 2

# Tentukan centroid awal
centroid_initial = df_encoded.sample(k, random_state=42).values

# Implementasi K-means
kmeans = KMeans(n_clusters=k, init=centroid_initial, n_init=1).fit(df_encoded.values)

# Visualisasi hasil
plt.scatter(df_encoded['Luas Lahan'], df_encoded['Produksi Padi'], c=kmeans.labels_, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='*', s=200, c='red', label='Centroids')
plt.xlabel('Luas Lahan')
plt.ylabel('Produksi Padi')
plt.title('Hasil Clustering K-means')
plt.legend()
plt.show()
```
